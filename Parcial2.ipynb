# Parcial 2

Se desea simular el comportamiento de un establecimiento comercial como una cola con servidor
único (Single-server Queue). El establecimiento ofrece servicio continuo entre las 8 am y las 4 pm (8
horas) y la disciplina de la cola es FIFO. Adicionalmente, se sabe que los tiempos entre llegadas
(minutos)siguen una distribución exponencial con λ=5/10 y los tiempos de servicio (minutos) siguen
una distribución exponencial con λ=6/10. 
Realice una simulación con 1000 réplicas y en cada réplica calcule lo siguiente:
- Número de trabajos.
- Estadísticos de trabajo.
- Estadísticos de tiempo.
Una vez realizadas las réplicas, calcule los intervalos de confianza de 95% para:
- Número de trabajos.
- Estadísticos de trabajo.
- Estadísticos de tiempo.

```python
from scipy.stats import expon, sem, t #importamos estas librerias para obtener los numeros random
from scipy import mean

TRABAJO=[0.0 for q in range(5)]
LLEGADAS=[0.0 for q in range(5)]
TASALL=[0.0 for q in range(5)]
SERVICIO=[0.0 for q in range(5)]
TSERVICIO=[0.0 for q in range(5)]
RETRASO=[0.0 for q in range(5)]
ESPERAPRO=[0.0 for q in range(5)]
TIEMPONODO=[0.0 for q in range(5)]
TIEMPOCOLA=[0.0 for q in range(5)]
TIEMPOSERVICIO=[0.0 for q in range(5)]

def confianza(valores): #Funcion para calcular los intervalos de confianza de 95%
  n = len(valores)
  m = mean(valores)
  std_err  = sem(valores)
  h = std_err*t.ppf((1+0.95)/2,n-1)
  start=m-h
  end=m+h
  return start,end

for q in range(1000): #se crea un for para la realizacion de las replicas

  TiempoEntreLLegada=[]
  TiempodeLLegada=[]
  TiempoServicio=[]
  i=0
  cont=0
  cont2=0
  cont6=0
  while(cont<480):#se coloca un while para generar los numeros aleatorios y calcular los tiempos de llegada sin pasasrse de las 8 horas
    TiempoEntreLLegada.append(expon.rvs(scale=2,size=1))
    TiempoServicio.append(expon.rvs(scale=1.6,size=1))
    cont+=TiempoEntreLLegada[i]
    TiempodeLLegada.append(cont)
    i=i+1
  a=len(TiempodeLLegada)
  c, d = [0.0 for x in range(a+1)], [0.0 for y in range(a+1)]
  W=[0.0 for y in range(a)]
  cont=0
  while(cont < a):# se obtiene los valores de tiempo que dura el nodo del servicio y el tiempo en que llega al servicio
    cont+=1
    if(TiempodeLLegada[cont-1] < c[cont-1]):
      d[cont-1] = c[cont-1] - TiempodeLLegada[cont-1]
    else:
      d[cont-1]=0.0
    c[cont] = TiempodeLLegada[cont-1] + d[cont] + TiempoServicio[cont-1]
    
  TRABAJO[q]=a



#se optiene las estadisticas de trabajo y de tiempo
  R=TiempodeLLegada[len(TiempodeLLegada)-1]/a
  LLEGADAS[q]=R

  tasaLL=1/R
  TASALL[q]=tasaLL

  cont4=0
  for v in range (len(TiempodeLLegada)):
	  cont4=cont4+TiempoServicio[v]
  S=cont4/a
  SERVICIO[q]=S

  tasaS=1/S
  TSERVICIO[q]=tasaS

  cont5=0
  for i in range (a):
	  cont5=cont5+d[i]
  D=cont5/a
  RETRASO[q]=D

  for i in range(a):
	  W[i]=d[i]+float(TiempoServicio[i])

  sumaW=0
  for i in range (a):
	  sumaW=sumaW+W[i]
  WR=sumaW/a
  ESPERAPRO[q]=WR

  I=(a/c[a-1])*WR
  TIEMPONODO[q]=I

  Q=(a/c[a-1])*D
  TIEMPOCOLA[q]=Q

  X=(a/c[a-1])*S
  TIEMPOSERVICIO[q]=X
# se mandan los datos a la funcion de "confianza" para imprimir el resultado del intervalo de confianza
print ("\nIntervalo de confianza del numero de trabajos: ", confianza(TRABAJO))

print ("\nIntervalo de confianza del promedio de llegadas : ", confianza(LLEGADAS))

print ("\nIntervalo de confianza del promedio tasa de llegadas : ", confianza(TASALL))

print ("\nIntervalo de confianza del promedio de servicios : ", confianza(SERVICIO))

print ("\nIntervalo de confianza del promedio tasa de servicios : ", confianza(TSERVICIO))

print ("\nIntervalo de confianza del promedio de retraso : ", confianza(RETRASO))

print ("\nIntervalo de confianza del promedio de espera : ", confianza(ESPERAPRO))

print ("\nIntervalo de confianza del promedio del tiempo en el nodo : ", confianza(TIEMPONODO))

print ("\nIntervalo de confianza del promedio de tiempo en la cola : ", confianza(TIEMPOCOLA))

print ("\nIntervalo de confianza del promedio de tiempo en servicio : ", confianza(TIEMPOSERVICIO))
```
